```python
import random

class Tiles:
    def __init__(self, name, price, rent):
        self.name = name
        self.price = price  # in millions
        self.rent = rent

class Board:
    def __init__(self):
        self.positions = {
            1: Tiles("GO", -2, 0),
            2: Tiles("Chicago", 8, 0.15),
            3: Tiles("SF", 10, 0.2),
            4: Tiles("Atlanta", 6, 0.1),
            5: Tiles("Irvine", 5, 0.1),
            6: Tiles("Buffalo", 3, 0.05)
        }
        self.p1 = Player(1)
        self.p2 = Player(2)
        self.p3 = Player(3)
        self.p4 = Player(4)


class Player:
    def __init__(self, id):
        self.id = id
        self.cash = 100
        self.properties = []
        self.position = 1

class GameMaster(Board):
    def __init__(self):
        super().__init__()
        self.p_id = 1
        self.current_player = self.p1
        self.players = [self.p1, self.p2, self.p3, self.p4]
        self.bank = Bank()
        self.owners = {}

    def switch_turn(self):
        self.p_id += 1
        if self.p_id > 4:
            self.p_id = 1
        self.current_player = self.players[self.p_id - 1]


    def roll_dice(self):
        return random.randint(1, 6)

    def move_player(self, player, roll):
        player.position += roll
        if player.position > len(self.positions):
            player.position -= len(self.positions)
            player.cash += 2  # Collect $2 for passing GO

    def interact(self):
        player = self.current_player
        roll = self.roll_dice()
        self.move_player(player, roll)
        tile = self.positions[player.position]

        print(f"Player {player.id} rolled a {roll} and landed on {tile.name}")

        if tile.price > 0:  #tile handling
            if tile not in self.owners:
              inp = input(f"P{player.id}, purchase {tile.name}? y/n")
              if(inp=='y'):
                if player.cash >= tile.price:
                    player.cash -= tile.price
                    player.properties.append(tile)
                    self.bank.receive_money(tile.price)
                    self.owners[tile]=player
                    print(f"Player {player.id} bought {tile.name} for {tile.price} million.")
                else:
                    print(f"Player {player.id} does not have enough cash to buy {tile.name}.")
            else:
                rent = tile.rent
                player.cash -= rent
                self.owners[tile].cash += rent
                print(f"Player {player.id} paid {rent} million to Player {self.owners[tile].id} in rent for landing on {tile.name}.")

        self.switch_turn()

class Bank:
    def __init__(self):
        self.cash = float('inf')

    def receive_money(self, amount):
        pass  # Infinite cash, so no need to track

game = GameMaster()

# game play
for _ in range(30):
    game.interact()
    print(f"Player 1 cash: {game.p1.cash}, Player 2 cash: {game.p2.cash}, Player 3 cash: {game.p3.cash}, Player 4 cash: {game.p4.cash}")
```
